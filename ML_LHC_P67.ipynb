```python
merged_Data = pd.merge(dpost,dataset5,on='post_id')
merged_Data
```

```python
l1=[]
unique = dict()
for i in merged_Data['tags']:
  #print(i)
  l1.append(len(i.split('>'))-1)
  tg=i.split('>')
  tg.pop()
  for j in tg:
    j=j.lower()
    if (j[1:] in unique):
      unique[j[1:]] += 1
    else:
      unique[j[1:]] = 1

print(unique)
```

```python
sortunique = sorted(unique, key=unique.get, reverse=True)
print(sortunique)
```

```python
sorted_unique = sorted(unique.items(), key=lambda kv: kv[1],reverse=True)
```

```python
import collections

sorted_dict = collections.OrderedDict(sorted_unique)
sorted_dict
```

```python
l2=[]
uniqueOverall = dict()
for i in dataset5['tags']:
  #print(i)
  l2.append(len(i.split('>'))-1)
  tg=i.split('>')
  tg.pop()
  for j in tg:
    j=j.lower()
    if (j[1:] in uniqueOverall):
      uniqueOverall[j[1:]] += 1
    else:
      uniqueOverall[j[1:]] = 1

print(uniqueOverall)
```

```python
sorted_uniqueOverall = sorted(uniqueOverall.items(), key=lambda kv: kv[1],reverse=True)
```

```python
import collections

sorted_dictOverall = collections.OrderedDict(sorted_uniqueOverall)
sorted_dictOverall
```

```python
for sub in sorted_dict:
    # checking if key present in other dictionary
    if sub in sorted_dictOverall:
        #sorted_dictOverall[sub]  = sorted_dictOverall.get(sub) - sorted_dict.get(sub)
        sorted_dictOverall[sub] = (sorted_dict.get(sub)/sorted_dictOverall.get(sub))*100
sorted_dictOverall
```

```python
from itertools import islice

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))
n_items = take(20, sorted_dictOverall.items())
n_items

top20 = collections.OrderedDict(n_items)
tags20 = list(top20.keys())
values20 = list(top20.values())
plt.figure(figsize=(25,15))
plt.plot(tags20,values20)
```

```python
ln=merged_Data['Create_Date']
ln2=merged_Data['Marked_Duplicate_Date']
type(ln[0])
```

```python
from datetime import datetime
Count_Day = []
for i in range(len(ln)):
  start=datetime.strptime(ln[i],"%Y-%m-%d")
  end= datetime.strptime(ln2[i],"%Y-%m-%d")
  Count_Day.append((end-start).days)
```

```python
for i in range(len(Count_Day)):
  Count_Day[i]=Count_Day[i]//365
Count_Day
```

```python
merged_Data['Years_To_Duplicate']=Count_Day
```

```python
freq = merged_Data.groupby(['Years_To_Duplicate']).size()
plt.figure(figsize=(30,35))
freq.plot.barh()
```
