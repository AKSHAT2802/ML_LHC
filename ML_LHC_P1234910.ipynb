```python
dataset2 = pd.read_csv('posts_short.csv')
print(dataset2)
```

```python
l1=[]
unique = dict()
for i in dataset2['tags']:
  #print(i)
  l1.append(len(i.split('>'))-1)
  tg=i.split('>')
  tg.pop()
  for j in tg:
    j=j.lower()
    if (j[1:] in unique):
      unique[j[1:]] += 1
    else:
      unique[j[1:]] = 1
print(l1)
dataset2['count']=l1
#dataset2
print(unique)
```

```python
freq = dataset2.groupby(['count']).size()
plt.figure(figsize=(25,15))
freq.plot.barh()
```

```python
sortunique = sorted(unique, key=unique.get, reverse=True)
print(sortunique)
```

```python
total = len(unique)
print("Total Number of Unique Tags ", total)
```

```python
top25=0
print("The top 25 Tags in Stack Overflow are : - ")
for i in sortunique:
  print(i)
  top25+=1
  if top25==25:
    break
```

```python
sorted_unique = sorted(unique.items(), key=lambda kv: kv[1],reverse=True)
```

```python
sorted_unique
type(sorted_unique)
import collections

sorted_dict = collections.OrderedDict(sorted_unique)
sorted_dict
```

```python
from itertools import islice

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))
n_items = take(25, sorted_dict.items())
n_items

top25 = collections.OrderedDict(n_items)
tags25 = list(top25.keys())
values25 = list(top25.values())
plt.figure(figsize=(25,15))
plt.plot(tags25,values25)
```

```python
que_count = 0
Total_count = 0
for i in dataset2['tags']:
  Total_count += 1
  temp = i.split('>')
  que_tag_count = len(temp)-1
  checked_tags = 0
  for j in temp:
    if j[1:] in tags500:
      checked_tags += 1
  if checked_tags == que_tag_count:
        que_count += 1
print(Total_count)
print(que_count)
```

```python
print("Percentage of Questions covered in top 500 Tags are : ",(que_count/Total_count)*100)
```
