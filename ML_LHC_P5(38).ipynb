```python
dataset1 = pd.read_csv('postLinks.csv')
print(dataset1)
```

```python
dataset1['new']=dataset1["creation_date"].str.split("-").str.get(0)+"-"+dataset1["creation_date"].str.split("-").str.get(1)
```

```python
dpost=dataset1[dataset1["link_type_id"]==3]
dpost
```

```python
aaa = dpost['new'].value_counts()
aaa
```

```python
plt.figure(figsize=(25, 15))
aaa.plot.barh()
```

```python
freq = dpost.groupby(['new']).size()
plt.figure(figsize=(30,35))
freq.plot.barh()
```

```python
type(freq)
```

```python
import collections

duplicates = collections.OrderedDict(freq)
```

```python
duplicates
```

```python
m_year = list(duplicates.keys())
total = list(duplicates.values())
plt.figure(figsize=(25,15))
plt.plot(m_year,total)
```

```python
freq = dataset1.groupby(['new']).size()
plt.figure(figsize=(30,35))
freq.plot.barh()
```

```python
# ratio w.r.t. previous month
ratio = []
for i in range(len(total)) :
    if(i == 0):
         ratio.append(total[0])
         continue
    ratio.append(int((total[i]/total[i-1])*100))
print(ratio)
ratio.pop(0)
m_year.pop(0)
```

```python

plt.figure(figsize=(30,30))
plt.plot(m_year,ratio)
```
